### 1. 다음은 2021년 국내 학생건강검사 표본조사 원자료 중 일부의 데이터이다. 데이터의 행과 열의 크기, 변수명, 변수 별 결측치의 개수, 변수 별 데이터 타입 등 기본 정보를 출력하시오.

```python
# [출처] 교육부, 학생건강정보센터, 학생건강검사 표본조사 원자료 (2021)
# https://www.data.go.kr/data/15100360/fileData.do?recommendDataYn=Y
from pandas import read_csv
df = read_csv('https://raw.githubusercontent.com/algoboni/pythoncodebook_visual1/main/data/1-4_studentHealth2021.csv')
df
```

```python
'''
데이터의 행과 열의 크기, 변수명, 변수 별 결측치의 개수, 변수 별 데이터 타입 등 기본 정보를 출력하시오.
'''
print(df.shape)
print(df.columns)
print(df.isna().sum())
print(df.dtypes)
```

### 2. 데이터프레임 df의 데이터 중 결측치가 있는 행을 삭제하고, 인덱스를 초기화 하시오. 적용 후, df의 행과 열의 크기를 출력하시오. (단, 기존의 인덱스는 데이터에서 제거한다.)

```python
df1 = df.dropna(axis=0).reset_index(drop=True)
df1.shape
```

### 3. 데이터프레임 df의 변수 중, '혈당(식전)(mg_dl)', '총콜레스테롤(mg_dl)', '충치치아_유무'의 컬럼명을 '식전혈당', '총콜레스테롤', '충치유무'로 변경하여 데이터프레임 df2를 생성하고, 첫 5개 행을 출력하시오.

```python
df2 = df1.copy()
df2 = df2.rename({'혈당(식전)(mg_dl)':'식전혈당', '총콜레스테롤(mg_dl)':'총콜레스테롤', '충치치아_유무':'충치유무'}, axis=1)
df2.head(5)
```

### 4. 데이터프레임 df2의unique 중, 범주형 변수인 '도시규모'와 '학교급'의 범주들을 출력하고, 도시 규모 혹은 학교급 순으로 수치형 변수로 값을 변환하여 해당 데이터프레임에 적용하시오. 적용 후 df2의 첫 5개 행을 출력하시오. (단, 도시 규모 혹은 학교급이 큰 순으로 3, 2, 1로 지정한다.)

```python
print(df2['도시규모'].unique())
print(df2['학교급'].unique())
```

```python
df2['도시규모'] = df2['도시규모'].map({c:i for i, c in enumerate(df2['도시규모'].unique())})
df2['학교급'] = df2['학교급'].map({c:i for i, c in enumerate(df2['학교급'].unique())})
df2.head(5)
```

### 5. 데이터프레임 df2의 변수 중, 범주형 변수인 '성별'과 '충치유무'의 범주들을 출력하고 수치형 변수로 값을 변환하여 해당 데이터프레임에 적용하시오. 적용 후 df2의 첫 5개 행을 출력하시오. (단, 남자 1, 여자 0으로 충치 유 1, 무 0, 검사안함은 -1로 지정한다.)

```python
print(df2['성별'].unique())
print(df2['충치유무'].unique())
```

```python
df2['성별'] = df2['성별'].map({c:i for i, c in enumerate(df2['성별'].unique())})
df2['충치유무'] = df2['충치유무'].map({c:i for i, c in enumerate(df2['충치유무'].unique())})
df2.head(5)
```

### 6. 데이터프레임 df2의 수치형 변수인 '키', '몸무게', '식전혈당', '총콜레스테롤', '허리둘레'를 소수점 첫째자리까지 반올림하고, '충치유무', '라면', '음료수', '패스트푸드', '육류', '과일', '아침식사'의 데이터 타입을 정수로 변경하여 적용하시오. 적용 후 df2의 첫 5개 행을 출력하시오. 

```python
nums = ['키', '몸무게', '식전혈당', '총콜레스테롤', '허리둘레']
df2[nums] = df2[nums].round(1)

ins = ['충치유무', '라면', '음료수', '패스트푸드', '육류', '과일', '아침식사']
df2[ins] = df2[ins].astype(int)
df2.head(5)
```

### 7. 데이터프레임 df2의 복사본인 df3를 생성하고, '생년월일'과 '건강검진일'로 응답 당시의 만 나이를 계산하여 '연령'이라는 파생변수를 생성하고 적용하시오. df3의 '생년월일'과 '건강검진일' 컬럼은 삭제한 후, df3의astype개 행을 출력하시오.

```python
import pandas as pd
df3 = df2.copy()
df3['건강검진일'] = pd.to_datetime(df3['건강검진일'].astype(int).astype(str))
df3['생년월일'] = pd.to_datetime(df3['생년월일'].astype(str))
df3['연령'] = df3['건강검진일'] - df3['생년월일']
df3['연령'] = (df3['연령'] / pd.Timedelta(days=365.2425)).astype(int)
df3 = df3.drop(['건강검진일','생년월일'],axis=1)
df3.head(3)
```

### 8. 데이터프레임 df3에서 '연령'이 0인 데이터를 삭제하고 데이터의 인덱스를 초기화 하시오. 적용 후, df3의 행과 열의 크기를 출력하시오. (단, 기존의 인덱스는 데이터에서 제거한다.)

```python
df3 = df3[df3['연령'] !=0].reset_index(drop=True)
df3.shape
```

### 9. 데이터프레임 df3의 '키', '몸무게'로 체질량지수를 계산하여 'BMI'라는 파생변수를 생성하고 적용한 후, df3의 첫 5개 행을 출력하시오. (단, 체질량지수는 몸무게(kg)를 키(m)의 제곱으로 나눈 값으로 소수점 첫째자리까지 반올림 한다. 본 데이터의 몸무게와 키의 단위는 각각 kg, cm임을 유의한다.) 

```python
df3['BMI'] = (df['몸무게'] / ((df['키']/100)**2)).round(1)
df3['BMI']
```

### 10. 데이터프레임 df3으로 학교급과 성별에 따른 키, 몸무게, 허리둘레, BMI의 평균을 출력하시오.

```python
df3.columns
```

```python
df3.groupby(['학교급','성별']).agg({c:'mean' for c in ['키', '몸무게', '허리둘레', 'BMI']})
```

### 11. 데이터프레임 df3의 'BMI'로 아래 기준에 따라 '체중구분'이라는 파생변수를 생성하여 적용하고, df3의 첫 5개 행을 출력하시오.  
[기준]
- 저체중: BMI 18.5 미만
- 정상 체중: BMI 18.5 이상 24.9 이하
- 과체중: BMI 25 이상 29.9 이하
- 비만: 30 이상

```python
import numpy as np
def dt(x):
    if x<18.5:return "저체중"
    elif x<24.9:return "정상 체중"
    elif x<29.9:return "과체중"
    elif x>30:return "비만"
    else : return np.NaN
df3['체중구분'] = df3['BMI'].apply(dt)
df3['체중구분'].head(5)
```

```python
df3['체중구분'].value_counts()
```

### 12. 데이터프레임 df3에서 체중구분에 따른 학생수와 그 비율을 데이터프레임으로 생성하여 출력하시오.

```python
df3.head(1)
```

```python
df4 = df3.groupby('체중구분').agg({'충치유무':'count'}).reset_index()
df4.columns = ['체중구분','학생수']
df4['비율'] = df4['학생수']/df4['학생수'].sum()
df4
```

### 13. 데이터프레임 df3으로 학교급, 성별, 체중구분에 따른 학생수를 출력하시오.

```python
df5 = df3.groupby(['학교급','성별','체중구분']).agg({'BMI':'count'},name='학생수').reset_index()
df5.columns = ['학교급','성별','체중구분','학생수']
df5
```

### 14. 데이터프레임 df3으로 학교급, 성별, 체중구분에 따른 BMI의 평균, 최솟값, 최댓값을 출력하시오.

```python
df6 = df3.groupby(['학교급','성별','체중구분']).agg({'BMI':['mean','min','max']}).reset_index()
df6
```
