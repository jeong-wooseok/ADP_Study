###  5.3.2
다음은 어느지역의 41208명의 생일을 월별로 기록한 자료이다. 월별 상대도수를 구하고 그것를 볼 때 각 달에 태어날 확률이 같다는 확률모형은 타당한지 결정하라

```python
import pandas as pd
birth = pd.DataFrame(
             {'탄생월':[f'{a}월' for a in range(1,13)],
             '인원':[3478, 3333, 3771, 3542, 3479,3304,3476,3295,3490,3331,3188,3321]
             })

birth['비중'] = birth['인원'] /  birth['인원'].sum()
birth
```

### 5.6.1

- 카이제곱 검정을 사용하여 월별 상대도수가 같은 확률 모형인지 평가

- 귀무가설 (H₀): 각 달에 태어날 확률이 같다 (즉, 월별 상대도수가 동일하다).
- 대립가설 (H₁): 각 달에 태어날 확률이 같지 않다 (즉, 월별 상대도수가 다르다).

- 단계
1. 카이제곱 통계량 계산: 관찰된 빈도와 기대 빈도 간의 차이를 제곱하여 합산한 후, 기대 빈도 나눔
2. 유의수준 설정: 유의수준 (일반적으로 0.05)을 선택하여 기각역을 설정합니다.
3. 기각 여부 결정: 계산한 카이제곱 통계량이 기각역을 넘으면 귀무가설을 기각하고, 그렇지 않으면 귀무가설을 채택합니다.

```python
from scipy.stats import chi2_contingency


# 기대 빈도 계산 (전체 인원 수 * 월별 상대도수)
total = 41208
expected_counts = birth['비중'] * total

# 카이제곱 검정 수행
chi2, p, dof, expected = chi2_contingency([birth['비중'] * total])

# 유의수준 0.05로 기각 여부 결정
alpha = 0.05
if p < alpha:
    conclusion = "귀무가설 기각: 월별 상대도수가 같지 않다."
else:
    conclusion = "귀무가설 채택: 월별 상대도수가 같다."

print("p-value = ", p, conclusion)
```

표본의 크기가 64인 자료에 대하여 모평균이 65보다 작다는 주장을 검정하고자 한다.σ=5일 때 다음 물음에 답하라

(1) 이 검정에 대한 기각역이 
일 때 1종 오류를 범할 확률?

```python
from scipy.stats import norm
import numpy as np
m, sigma = 65, 5
n = 64
x = 64
se = (x - m) / (sigma / n **0.5)
res = norm.cdf(se)
print(f'제1종 오류를 범할 확률: {res:.3f}')
```
